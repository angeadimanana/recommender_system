```python
import sys, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append('/content/drive/MyDrive/recommender_system/src')
```

```python
movies = pd.read_csv("data/ml-32m/movies.csv")
ratings = pd.read_csv("data/ml-32m/ratings.csv")

```

```python
rating_counts = ratings['rating'].value_counts().sort_index()

plt.figure()
plt.bar(rating_counts.index, rating_counts.values, color='cornflowerblue', edgecolor='black', width=0.3)

plt.xlabel("Rating")
plt.ylabel("Number of movies")
plt.xticks(rating_counts.index)
plt.grid(alpha=0.3, axis='y')
# path = "figure_saved"
# name_file = f"Histogram of ratings.pdf"

# os.makedirs(path, exist_ok=True)

# plt.savefig(os.path.join(path, name_file), format='pdf',bbox_inches="tight")

plt.show()

```

```python
movies['genre_list'] = movies['genres'].apply(lambda x: x.split('|'))
genres = [g for sublist in movies['genre_list'] for g in sublist]

from collections import Counter
genre_counts = Counter(genres)

plt.figure()
plt.barh(list(genre_counts.keys()), list(genre_counts.values()), color='lightcoral')
plt.xlabel("Number of movie")
plt.ylabel("Genre")
plt.grid(alpha=0.3, axis='x')
# path = "figure_saved"
# name_file = f"genre_movie_distribution.pdf"

# os.makedirs(path, exist_ok=True)

# plt.savefig(os.path.join(path, name_file), format='pdf',bbox_inches="tight")


plt.show()
```
